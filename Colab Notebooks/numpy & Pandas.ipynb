```python
df.job_posted_date = pd.to_datetime(df.job_posted_date)
df.info()
```

```python
df['job_posted_month'] = df.job_posted_date.dt.month
df.sort_values('job_posted_date', inplace=True)
```

```python
df
```

```python
df.drop(labels="salary_hour_avg", axis = 1, inplace=True)
```

```python
df.info()
```

```python
df.dropna(subset=['salary_year_avg'], inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['salary_year_avg'].idxmin()
df['job_title_short'].value_counts()
df.groupby('job_title_short')['salary_year_avg'].agg(['min','max','median'])
df['job_country'].isin(['North Korea']).any()
```

```python
kenyan_jobs = df[df['job_country'] == "Kenya"]
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,5,0.1)
y = np.sin(x)
plt.plot(x,y)
```

```python
date_counts = df.job_posted_date.value_counts()
date_counts = date_counts.sort_index()
month_counts = df.job_posted_month.value_counts();
month_counts = month_counts.sort_index()
```

```python
month_counts
```

```python
plt.plot(month_counts.index, month_counts.values)
plt.show()
```

```python
job_counts = df.job_title_short.value_counts()
job_counts = job_counts.sort_values(ascending = True)
plt.barh(job_counts.index, job_counts)
```

```python
plt.bar(job_counts.index, job_counts.values)
plt.title("Postings By Job Title")
plt.ylabel("Count Of Job Postings")
plt.xticks(rotation = 45, ha = "right")
plt.show()
```

```python
job_counts = job_counts.sort_values(ascending = False)
job_counts.plot(kind="bar")
plt.title("Postings By Job Title")
plt.ylabel("Count Of Job Postings")
plt.xlabel("")
plt.xticks(rotation = 45, ha = "right")
plt.show()
```

```python
job_salary = df.groupby("job_title_short")["salary_year_avg"].median().sort_values()
```

```python
job_salary.plot(kind="barh")
plt.xlabel('Salary ($USD)')
plt.ylabel('')
plt.title('Median salary by Job Title')
plt.show()
```
